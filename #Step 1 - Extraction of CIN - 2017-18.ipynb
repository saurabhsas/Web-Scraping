```python
#---------------------------------------------------------------------------
#Step 1 : Extract CIN(s) for Listed Companies of financial year 2017-18                    
#---------------------------------------------------------------------------
```

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen
```

```python
#Extract data as a table with pandas
#separate = pd.read_html('https://www.csr.gov.in/companylist.php?csr_spent_range=&state=&district=&mact=&year=FY%202019-20&compCat=&listing=Listed')
#new = pd.concat([separate[0]])
#new.head()
```

```python
html = urlopen("https://www.csr.gov.in/companylist.php?csr_spent_range=&state=&district=&mact=&year=FY%202017-18&compCat=&listing=Listed")
soup = BeautifulSoup(html.read(),'lxml')
links = []
for link in soup.find_all('a'):
    links.append(link.get('href'))
```

```python
links = links[61:] #removing extra list element
new_links = [x[-21:] for x in links] #extraction of las 21 characters for CIN
new_links = new_links[:2113] #keep only the CINs for 2808 companies

#Creating a DataFrame for listed companies
df1=pd.DataFrame({"CIN":new_links})
df1.count()
```

```python
#---------------------------------------------------------------------------
#Step 2 : Extract CIN(s) for Un-Listed Companies of financial year 2017-18                   
#---------------------------------------------------------------------------
```

```python
html = urlopen("https://www.csr.gov.in/companylist.php?csr_spent_range=&state=&district=&mact=&year=FY%202017-18&compCat=&listing=Unlisted")
soup = BeautifulSoup(html.read(),'lxml')
links = []
for link in soup.find_all('a'):
    links.append(link.get('href'))
```

```python
links = links[61:] #removing extra list element
new_links = [x[-21:] for x in links] #extraction of las 21 characters for CIN
new_links = new_links[:19404] #keep only the CINs for 2808 companies

#Creating a DataFrame for Un-listed companies
df2=pd.DataFrame({"CIN":new_links})
df2.count()
```

```python
#---------------------------------------------------------------------------
#Step 3 : Append Listed and Un-listed CIN                    
#---------------------------------------------------------------------------
df = df1.append(df2,ignore_index = True)
df.count()
```

```python
print(df.head(2),"\n",df.tail(2))
```
