```python
#---------------------------------------------------------------------------
#Step 1 : Data Import for Company Names for financial year 2016-17                    
#---------------------------------------------------------------------------
import pandas as pd
fy_18_19 =pd.read_excel(r"C:\Users\saura\Desktop\CSR\Data From National CSR Portal (Dec 2021)\CSR - Company CIN (2018-2019).xlsx","Financial Year 2018-2019", header=0) 
fy_18_19.count()
```

```python
fy_18_19.head()
```

```python
#-----------------------------------------------------------------------------------
#Step 2 : Web Scraping for variables from Table 2:
#CIN	csr_project	development_Sector	state_name	district	
#project_amount_outlay	amount_spent	mode_of_implementation	feedback
#-----------------------------------------------------------------------------------
from bs4 import BeautifulSoup
import requests
import warnings
warnings.filterwarnings('ignore')
import sys
import time
```

```python
fy_18_19_cin = fy_18_19.iloc[:,]
fy_18_19_cin.count()
```

```python
fy_18_19_new=fy_18_19_cin["CIN"].values.tolist()
print(fy_18_19_new)
```

```python
ds_new2=pd.DataFrame()
df2=pd.DataFrame()
```

```python
count = 0
retries = 1
for cin in fy_18_19_new:
    try:
        url = "https://www.csr.gov.in/companyprofile.php?year=FY+2018-19&CIN="+cin
        r = requests.get(url, timeout=30)
        #print(r,url)
        data = r.text
        new_data = r.text[data.find("CSR Details of 2018-19"):data.find("End of CSR Details of 2018-19")]
        soup = BeautifulSoup(new_data)
        count=count+1
        print(count)
        
        table2 = soup.find_all('table',class_="table")[1]
        #print(table2)
        rows = table2.find_all('tr')[1:]
        #print(rows)
        #Generate lists
        A=[         ]
        B=[         ]
        C=[         ]
        D=[         ]
        E=[         ]
        F=[         ]
        G=[         ]
        H=[         ]
        I=[         ]
        for row in table2.find_all("tr"):
            cells = row.find_all('td')
        
    #       length=len(cells)
    #       print(length)
            #states=row.findAll('th') #To store second column data
            if len(cells)==9: #Only extract table body not heading
                A.append(cells[0].find(text=True))
                B.append(cells[1].find(text=True))
                C.append(cells[2].find(text=True))
                D.append(cells[3].find(text=True))
                E.append(cells[4].find(text=True))
                F.append(cells[5].find(text=True))
                G.append(cells[6].find(text=True))
                H.append(cells[7].find(text=True))
                I.append(cells[8].find(text=True))
                #print(A)
            df2 = pd.DataFrame(A,columns=['S.No'])
            df2['financial_year'] = '2018-2019'
            df2['CIN'] = cin
            df2['csr_project']=B
            df2['development_Sector']=C
            df2['state_name']=D
            df2['district']=E
            df2['project_amount_outlay']=F
            df2['amount_spent']=G
            df2['mode_of_implementation']=H
            df2['feedback']=I
            #print(df2)
            #df2.set_index('S.No', inplace = True)
        f1 = [ds_new2,df2]
        ds_new2 = pd.concat(f1)
    except Exception as e:
        wait = retries * 30
        sys.stdout.flush()
        time.sleep(wait)
        retries += 1
#ds_new2.count()
```

```python
Table_final1 = ds_new2
Table_final1.head()
```
